```python
# data type
# list
my_list = [1, 2, 3, 4, 5]

# Methods 
print(my_list)
my_list.append(6)  # Adds an element to the end of the list
print(my_list)  # Output: [1, 2, 3, 4, 5, 6]

my_list.insert(2, 10)  # Inserts 10 at index 2
print(my_list)  # Output: [1, 2, 10, 3, 4, 5, 6]

my_list.remove(3)  # Removes the first occurrence of 3
print(my_list)  # Output: [1, 2, 10, 4, 5, 6]

my_list.pop()  # Removes and returns the last element
print(my_list)  # Output: [1, 2, 10, 4, 5]

# properties
#ordered and mutable
my_list[2]= 20  # Modifies the element at index 2
print(my_list)  # Output: [1, 2, 20, 4, 5]

# heterogeneous & duplicates allowed
zen=[1,2,3,4,5,1,2,3,"ai",True,False]
print(zen)  # Output: [1, 2, 3, 4, 5, 1, 2, 3, 'ai', True, False]
```

```python
# var 1=45
# function

class Car:     #human:
    # attributes  color, model, year       color,eye, nose
    
    def __init__(self, color, model, year):   
        self.color = color
        self.model = model
        self.year = year

    # methods  start stop play                move walk talk
    def start(self):               
        print("car is starting")

    def stop(self):
        print("car is stopping")

    def play(self, song="Despacito"):
        print(f"Playing {song}")

    
bmw = Car("red", "X5", 2020)
```

```python
# aryan
#create a class box , having some attributes  and placing methids (stack, push , pop ,append)
class Box:
        def __init__(self, size, shape, color, volume):
                self.size = size
                self.shape = shape
                self.color = color
                self.volume = volume
        def stack(self):
                print(f"Rohan is stacking a box of color: {self.color}, shape: {self.shape} , size : {self.size} and volume : {self.volume}")
        def push(self):
                print(f"The {self.color} {self.shape} box is pushing.")
        def pop(self):
                print(f"The {self.color} {self.shape} box is popping.")
md = Box("large", "square", "red", 100)
md.stack()
md.push()
md.pop()
```

```python
# WHILE LOOP USE FOR AND WHILE
while True:
    user_input = input("Enter a number (or 'q' to quit): ")
    if user_input.lower() == 'q':
        break
    try:
        number = int(user_input)
        print(f"You entered: {number}")
    except ValueError:
        print("Please enter a valid number.")
```
