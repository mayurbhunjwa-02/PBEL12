# Confusion Matrix

It is small table That actual labels with predicted labels.

1. True Positive : Predicted positive , actually positive

2. True Negative : Predicted negatiove , actually negative

3. False Positive : Predicted positive actually negative

4. False Negative : predicts negative actually positive

Examples: Email is spam Actual spam, predicted spam = TP

    - Actual not spam, predicted spam = FP

    - Actual spam, predicted not spam = FN

    - Actual not spam, predicted not spam = TN

# Accuracy =

Accuracy tells us model correct overall

Example: if 100 emails, the model gets 90 right, accuracy 90%

Mislead :

- accuracy can look high even when the model is poor on the important class .

- Example: if only 1 out of 100 emails is spam. a model that always says not spam not spam ->>>> Acuracy 99%

- useless for finding spams!
