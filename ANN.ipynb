```python
import tensorflow as tf
from tensorflow import keras
from matplotlib import pyplot as plt
import numpy as np
import matplotlib.image as img
```

```python
(X_train,Y_train), (X_test,Y_test)=keras.datasets.mnist.load_data()
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
X_train_flattened = X_train.reshape(len(X_train),28*28)
X_test_flattened = X_test.reshape(len(X_test),28*28)
```

```python
X_train_flattened.shape
```

```python
X_train_flattened[0]
```

```python
model = keras.Sequential([
    keras.layers.Dense(10,input_shape=(784,), activation='sigmoid')
])
model.compile(
    optimizer = 'adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model.fit(X_train_flattened,Y_train,epochs=5)
```

```python
model.evaluate(X_test_flattened,Y_test)
```

```python
Y_test[0]
```

```python
y_predicted = model.predict(X_test_flattened)
y_predicted[0]
```

```python
plt.matshow(X_test[0])
```

```python
np.argmax(y_predicted[0])
```

```python
y_predicted_labels = [np.argmax(i) for i in y_predicted]
```

```python
y_predicted_labels[:5]
```

```python
cm = tf.math.confusion_matrix(labels=Y_test,predictions=y_predicted_labels)
cm
```

```python
import seaborn as sns
plt.figure(figsize=(10,7))
sns.heatmap(cm, annot=True, fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model1= keras.Sequential([
    keras.layers.Dense(100,input_shape=(784,),activation='relu'),
    keras.layers.Dense(10,activation='sigmoid')
])
model1.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']

)
model1.fit(X_train_flattened,Y_train,epochs=5)
```

```python
model1.evaluate(X_test_flattened,Y_test,)
```

```python
y_predicted1 = model1.predict(X_test_flattened)
y_predicted_labels1 = [np.argmax(i) for i in y_predicted1]
cm1 = tf.math.confusion_matrix(labels=Y_test,predictions=y_predicted_labels1)

plt.figure(figsize=(10,7))
sns.heatmap(cm1,annot=True,fmt='d')
plt.xlabel('Predicted')
plt.ylabel('Truth')
```

```python
model12 = keras.Sequential([
    keras.layers.Flatten(input_shape=(28,28)),
    keras.layers.Dense(100,input_shape=(784,),activation='relu')
])
model12.compile(
    optimizer='adam',
    loss='sparse_categorical_crossentropy',
    metrics=['accuracy']
)
model12.fit(X_train,Y_train,epochs=5)
```
