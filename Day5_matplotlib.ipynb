```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd 
```

```python
tips = pd.read_csv("Data/tipsdataset.csv")
```

```python
##univariate, bivariate, multivariate ,# 1 col age
#biv: 2 col num-num, num-cat, cat-cat
#multi all data more than 3
```

```python
# univarite
plt.plot(tips['total_bill'])
plt.xlabel("EXAMPLE")
plt.title("PLOT")
plt.show()
```

```python
plt.scatter(tips['size'],tips['total_bill'],s=tips['total_bill'])
plt.title("scatter")
plt.show()
```

```python
#plt.pie() #cat using value count
#plt.pie(tips['total_bill'])
#plt.pie(tips['gender'].value_counts())

import matplotlib.pyplot as plt

plt.pie(
    tips['gender'].value_counts(),
    labels=tips['gender'].value_counts().index,
    autopct='%1.1f%%'
)
plt.show()
```

```python
plt.hist(tips['tip'], bins=4) #univariate num [1,20,30,40,50] div data 4 parts 1-100  1: 1:24 2:25-49 3,4
plt.show()  
```

```python
sns.relplot(
    data=tips,
    x="tip",
    y="size",
    col="gender",
    hue="gender"
)
```

```python
#create a visualization
sns.relplot(
    data=tips,
    x="total_bill",
    y="tip",
    hue="smoker",
    style="smoker",
    size="size"
)
```

```python
sns.histplot(data=tips,x="total_bill",hue="time",kde=True)
sns.histplot(data=tips,x="tip",hue="time",kde=True)
```

```python
sns.displot(data=tips, x="total_bill", col="time", kde=True)
sns.displot(data=tips, x="tip", col="time", kde=True)
```
