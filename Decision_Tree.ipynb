# **Understand structure of our decision tree**

### **Tree**

- Real Tree : it has -> Roots -> trunk -> branches -> leaves 

- Decision Tree : Roots -> Questions -> More Questions -> Final answers

- A decision tree grows exactly like a natural tree. Only difference is that instead of growing leaves it grows decisions. 

#### **Parent Node** : 

- Parent node createes one or more child nodes 

- Salary ( parents node ) -> yes or noo -> child node 


#### **Child Node** : 

- Every node Created from another node is called a child node.

Root node -> Internal node -> branch -> child node -> leaf node 

- This entire structure is called a Decision Tree!

### **Pure Vs Impure nodes**

- Basket A : You have apple, apple, apple, apple
- Pure

- Basket B : apple, banana, apple, orange, banana, apple
- Impure

- Tree loves pure nodes

- Tree hate impure nodes

- The goal is simple convert impure -> less impure -> almost pure -> completely pure 

- Decision tree grow by increasing purity. 

### ***Entropy*** 

- Imagine you are playing a guessing game.
- You have a box
- Inside : Dog or Cat
- Whats inside ?
- DOG.....
- You already know the answer
- THERE IS NO UNCERTAINITY

- Imagine another box 
- It can contain
- Inside : Cat or Dog or Rabbit or House or Fish or Bird
- Can you guess norw ?
- Because there are vmany possibilities
- The incertainity has increased
- This uncertainity is called    ******Entropy******

- Entropy measures " How much confusion or uncertainity exists in the data "
- Less confusion -> Low Entropy 
- More confusion  -> High Entropy 
- Decision tree always try to Reduce confusion

**Ex - Suppose there are 55 students**

- 55 passed
- 0 failed 
- Result ?
Entropy = Very Low

**Ex - 55 Students**

- 27 passed 
- 28 failed
- Prediction become difficult
- Entropy = High

**Ex - 55 students**

- 53 passed
- 2 failed
- Predict : Passed
- Entropy : Mixed data -> Low

****Note :****

- More mixed data -> High Entropy 
- More organized Data -> Lower Entropy

## **Gini Index ( The fast Decision maker )**

- Imagine there are two judges in a competition.

- Judge A :

- Think very carefully 

- Check every rule 

- Overall calculate everything 

- Finally gives the winner

- Judge B:

- Looks quickly

- Makes a fast decision.

- Still gives almost the same winner.

- Who finishes first ?

- Judge B

- This is exactly the difference between entropy and Gini Index.

- Both measure Impurity
- but Gini Index computationally faster.
- That is why the CART algorithm uses GINI INDEX.

### **Gini Index ?**

- Gini Index measures " How impure or mixed a dataset is "

- More mixed data ---->>> Higher Gini 

- More Pure data -->>> Lower Gini

- **Decision tree always try to Reduce GINI**

- Entropy ask " How much information is missing ? "

- GIni ask " How likely am i to classify a incorrectly ? "

- Both reached almost the same decision.

- But Gini requires fewer calculation.

- Therefore, Gini is faster.

### **Baskets :**

- **Basket A :** Apple, Apple, Apple, Apple

- Only APPLE

- Mistake ?

- Almost NEVER

- Gini = 0 

- Perfactly Pure.

- **Basket B :** Apple, Apple, Banana, banana, Apple, Banana

- Half Apples

- Half Banana

- Now mistakes can be possible

- Gini Higher

- **Bakest C :** Apple, Banana, Grapes, Watermelons, Oranges

- Everything mixed

- Prediction becomes difficult

- Highest Impurity

- Binary classification :

- gini = 1 - ( p1^2 + p2^2 )
- p1 = probability of class 1
- p1 = probability of class 2

- for multiple classes
- gini = 1 - $\sum$(pi^2) 

        # $\sum$ for submission symbol

- Dataset : 10 students 
- Pass - 10 
- Fails - 0

- Probability of pass = 10/10 = 1
- Probability of fail = 0/10 = 0

- Gini = 1 - ( 1 + 0 )
- Gini = 0
- Perfectly pure dataset

- Pass = 5
- Fail = 5
- Probability of pass = 0.5
- Probability of fail = 0.5
- Gini = 1 - ( 0.25 + 0.25 )
- Gini = 0.5
- Maximum Gini for binary classification 

- Customer 100
- Purchased - 80 
- Not purchased - 20 probabilty of purchase - 0.8 probabilty of unpurchased - 0.2

Gini = 1 - ( 0.64 + 0.04 ) gini = 0.32

dataset is Fairly pure

```python
def gini (p,n):
    t = p + n
    p1 = p/t
    p2 = n/t

    g = 1 - ( p1**2 + p2**2 )
    return g 

print("Pure Gini : ", gini(10,0))
print("Balanced Gini : ", gini(5,5))
print(" 80 - 20 Gini : ", round(gini(8,2),3))
```

#### **Basic  Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from sklearn.model_selection import train_test_split # Suffle + train and test, split
from sklearn.datasets import load_breast_cancer
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report, precision_score, recall_score, f1_score, classification_report
```

```python
from sklearn.tree import DecisionTreeClassifier
from sklearn.tree import plot_tree
```

### **Load Dataset :**

```python
cancer = load_breast_cancer()
df = pd.DataFrame(cancer.data, columns=cancer.feature_names)
df ["Target"] = cancer.target
```

```python

df.head(10)
```

### **Explore Data**

```python
print("shape:", df.shape)
```

```python
print(df.info())
```

```python
print(df['Target'].value_counts())
```

```python
X = df.drop('Target', axis=1)
y = df["Target"]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=42)
```

```python
print("training shape:", X_train.shape)
```

```python
print("Testing Shape:", X_test.shape)
```

## **Decision Tree Model Create**

```python
model = DecisionTreeClassifier(
    criterion = "gini",
    random_state=42
)
```

#### **Train Model** 

```python
model.fit(X_train, y_train)
```

#### **Predict**

```python
y_pred = model.predict(X_test)
```

```python
comparision = pd.DataFrame({
    'Actual':y_test.values,
    "predict": y_pred
})
```

```python
comparision.head(10)
```

#### **Accuracy**

```python
accuracy = accuracy_score(y_test,y_pred)
print("Acc:",round(accuracy,2))
```

#### **Precision, Recall, f1**

```python
Precision = precision_score(y_test, y_pred)
print ("Precision:",round(Precision,2))

Recall = recall_score(y_test,y_pred)
print("Recall:",round(Recall,2))


f1 = f1_score(y_test,y_pred)
print(f1)
```

#### **Confusion Matrix**

```python
cm = confusion_matrix(y_test,y_pred)
print(cm)
```

```python
plt.figure(figsize=(6,5))
sns.heatmap(cm,annot=True, cmap="Blues", fmt= 'd',xticklabels=['Maligant',"benign"], yticklabels=['maligant', "benign"])
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title("Decision Tree confusion matrix")
plt.show()
```

#### **Classification Report**

```python
print(classification_report(y_test,y_pred))
```

```python
plt.figure(figsize=(20,10))

plot_tree(
    model,
    feature_names=X.columns,
    class_names=['Malignant', 'Benign'],
    fontsize=9,
    rounded=True,
    filled=True
)

plt.show()
```
